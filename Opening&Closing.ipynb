```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('openingTest.PNG', 0)
kernel = np.ones((5,5),np.uint8)
```

```python
opening = cv2.morphologyEx(img1, cv2.MORPH_OPEN, kernel)
```

```python
print(opening)
```

```python
res = np.hstack((img1, opening))
plt.imshow(res, cmap='gray')
plt.show()
```

```python
img2 = cv2.imread('closingTest.PNG', 0)
```

```python
closing = cv2.morphologyEx(img2, cv2.MORPH_CLOSE, kernel)
```

```python
print(closing)
```

```python
res2 = np.hstack((img2, closing))
plt.imshow(res2, cmap='gray')
plt.show()
```
