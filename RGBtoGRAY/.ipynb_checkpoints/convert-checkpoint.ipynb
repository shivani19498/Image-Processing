```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
img = mpimg.imread('images.png')
gray = rgb2gray(img)
```

```python
plt.imshow(img, cmap='gray')
```

```python
plt.imshow(gray, cmap=plt.get_cmap('gray'))
plt.show()
```
